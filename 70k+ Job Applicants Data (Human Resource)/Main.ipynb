```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("HRdata.csv")
df.head(5)
```

```python
data = df.drop(columns=['Unnamed: 0'])
```

```python
data = data.dropna()
```

```python
data['Age'] = data['Age'].astype('category')
data['EdLevel'] = data['EdLevel'].astype('category')
```

```python
data['Employment'] = data['Employment'].astype(int)
data['YearsCode'] = data['YearsCode'].astype(int)
data['YearsCodePro'] = data['YearsCodePro'].astype(int)
data['ComputerSkills'] = data['ComputerSkills'].astype(int)
data['Employed'] = data['Employed'].astype(int)
```

```python
data.info()
```

```python
data.head()
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='Age', data=data)
plt.title('Distribution of Age Groups')
plt.xlabel('Count')
plt.ylabel('Age Group')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='EdLevel', data=data)
plt.title('Distribution of Education Levels')
plt.xlabel('Count')
plt.ylabel('Education Level')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='Gender', data=data)
plt.title('Distribution of Gender')
plt.xlabel('Count')
plt.ylabel('Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='MainBranch', data=data)
plt.title('Distribution of Main Branch of Work')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='Employed', data=data)
plt.title('Distribution of Employment Status')
plt.xlabel('Count')
plt.ylabel('Employed')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='YearsCode', y='PreviousSalary', data=data)
plt.title('Relationship between Years of Coding and Previous Salary')
plt.xlabel('Years of Coding')
plt.ylabel('Previous Salary')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['PreviousSalary'], kde=True)
plt.title('Distribution of Previous Salary')
plt.xlabel('Previous Salary')
plt.ylabel('Count')
plt.show()
```

```python
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
corr_matrix = data[numeric_cols].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numeric Variables')
plt.show()
```
